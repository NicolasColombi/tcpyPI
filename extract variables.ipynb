```python
# select variables
q = monthly_atm.sphum
t = monthly_atm.temp
msl = monthly_atm.slp
p = monthly_atm.level
sst = sst_monthly
```

```python
# Rename dimensions or coordinates in the entire dataset
ds_renamed = combined_ds.rename({
    'time': 'month',
    'level': 'p'
})
```
